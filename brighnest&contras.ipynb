```python
import numpy as np #import lib numpy
import cv2 #import lib cv2
import matplotlib.pyplot as plt #import lib matplotlib
```

```python
img = cv2.imread('orange.jpg') #membaca dan mengambil gambar dari folder
```

```python
img_height= img.shape[0] #mengambil tinggi gambar (0)
img_width= img.shape[1] #mengambil lebar gambar (1)
img_channel= img.shape[2] #mengambil channel warna gambar (2)
img_type= img.dtype #menginisiasi tipe gambar
```

```python
img_brightness = np.zeros(img.shape, dtype= np.uint8) # untuk meng kosongkan array yg sudah ada pada gambar
```

```python
def cerah(nilai): # fungsi brighter dengan parameter nilai
    for y in range (0, img_height): # Looping untuk tinggi gambar 
        for x in range (0, img_width): # Looping untuk lebar gambar
            red= img[y][x][0]
            green= img[y][x][1]
            blue= img[y][x][2]
             # untuk mendapatkan informasi piksel gambar dari RGB
            gray= (int(red) + int(green)+ int(blue))/3 # menghitung nilai rata-rata untuk mendapatkan nilai keabuan
            gray += nilai
            if gray > 255:
                gray = 255
                # nilai batas
            img_brightness[y][x] = (gray, gray, gray) # untuk menampung nilai baru
```

```python
cerah(-100) # memanggil fungsi brighter dengan nilai -100
plt.imshow(img_brightness) # menampilkan gambar
plt.title('kecerahan -100') # menampilkan title gambar
plt.show() # menampilkan 

cerah(25) # memanggil fungsi brighter dengan nilai 100
plt.imshow(img_brightness)  # menampilkan gambar
plt.title('kecerahan 100') # menampilkan title gambar
plt.show() # menampilkan
```

```python
img_rgbbright = np.zeros(img.shape, dtype=np.uint8) # untuk meng kosongkan array yg sudah ada pada gambar
```

```python
def rgbcerah(nilai): # fungsi rgbbrighter dengan parameter nilai
    for y in range(0, img_height): # Looping untuk tinggi gambar 
        for x in range(0, img_width): # Looping untuk Lebar gambar 
            red = img[y][x][0] # untuk mendapatkan informasi piksel gambar dari red
            red += nilai
            if red > 255: # nilai batas
                red = 255
            if red < 0: # nilai batas
                red = 0
            green = img[y][x][1] # untuk mendapatkan informasi piksel gambar dari green
            green += nilai
            if green > 255: # nilai batas
                green = 255
            if green < 0: # nilai batas
                green = 0
            blue = img[y][x][2] # untuk mendapatkan informasi piksel gambar dari blue
            blue += nilai
            if blue > 255: # nilai batas
                blue = 255
            if blue < 0: # nilai batas
                blue = 0
            img_rgbbright[y][x] = (red, green, blue) #  untuk menampung nilai baru
```

```python
rgbcerah(-100) # memanggil fungsi rgbbrighter dengan nilai -100
plt.imshow(img_rgbbright) # menampilkan gambar
plt.title("Brightness -100") # menampilkan title gambar
plt.show() # menampilkan plot

rgbcerah(100) # memanggil fungsi rgbbrighter dengan nilai 100
plt.imshow(img_rgbbright) # menampilkan gambar
plt.title("Brightness 100") # menampilkan title gambar
plt.show() # menampilkan plot
```

```python
img_contrass= np.zeros(img.shape, dtype=np.uint8) # untuk meng kosongkan array yg sudah ada pada gambar
```

```python
#contrass auto level

img_autocontrass= np.zeros(img.shape, dtype=np.uint8)  # untuk meng kosongkan array yg sudah ada pada gambar
```

```python
def autocontrass():  # fungsi contrasss
    xmax=300 # mendefinisikan variabel xmax dengan 255
    xmin=0 # mendefinisikan variabel xmin dengan 0
    d = 0  # mendefinisikan variabel d dengan 0

    for y in range(0, img_height):  # Looping tinggi gambar 
        for x in range(0, img_width):  # Looping lebar gambar 
            red=img[y][x][0]
            green=img[y][x][1]
            blue=img[y][x][2]
             # untuk mendapatkan informasi piksel gambar dari RGB
            gray=(int(red) + int(green) + int(blue))/3  # menhitung nilai rata-rata untuk mendapatkan nilai keabuan
            if gray < xmax:
                xmax = gray
            if gray > xmin:
                xmin = gray
                 # nilai batas
        d = xmax - xmin  # menghitung selisih xmin-xmax dan menyimpannya di variable d
        for y in range(0, img_height):
            for x in range(0, img_width):
                red= img[y][x][0]
                green= img[y][x][1]
                blue= img[y][x][2]
                gray= (int(red) + int(green) + int(blue))/3
                gray=int(float(255/d)*(gray-xmax))  # rumus menghitung nilai keabuan baru
                img_autocontrass[y][x] = (gray, gray, gray) #menampung nilai baru
```

```python
autocontrass()  # memanggil fungsi auto contrass
plt.imshow(img_autocontrass) # menampilkan gambar
plt.title('autocontrass') # menampilkan title
plt.show() # menampilkan plot
```
